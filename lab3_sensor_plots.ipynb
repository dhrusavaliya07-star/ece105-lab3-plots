```python
# Generate simulated temperature readings from two sensors.
# Sensor A: mean 25 C, std 3 C, 200 readings.
# Sensor B: mean 27 C, std 4.5 C, 200 readings.
#can Also generate 200 timestamps uniformly from 0 to 10 seconds.
# Use np.random.default_rng with a seed = last 4 digits of your Drexel ID.
import numpy as np
import matplotlib.pyplot as plt

# Replace 1234 with the last 4 digits of your Drexel ID for reproducibility
rng = np.random.default_rng(1234)

n_readings = 200
timestamps = rng.uniform(0.0, 10.0, size=n_readings)
sensor_a = rng.normal(loc=25.0, scale=3.0, size=n_readings)
sensor_b = rng.normal(loc=27.0, scale=4.5, size=n_readings)
```

```python
# Scatter plot: sensor reading (y-axis) vs timestamp in seconds (x-axis).
# Color Sensor A points blue and Sensor B points orange.
# Add axis labels with units, a title, and a legend.
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.scatter(timestamps, sensor_a, s=40, c='tab:blue', alpha=0.7, label='Sensor A', edgecolors='k')
plt.scatter(timestamps, sensor_b, s=40, c='tab:orange', alpha=0.7, marker='s', label='Sensor B', edgecolors='k')

plt.xlabel('Time (s)')
plt.ylabel('Temperature (°C)')
plt.title('Scatter Plot: Temperature vs Time for Two Sensors')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Overlaid histogram of Sensor A and Sensor B temperature distributions.
# Use 30 bins, alpha=0.5 for transparency so both distributions are visible.
# Add vertical dashed lines at each sensor's mean.
# Include a legend labeling each sensor.
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
# Use 30 bins and semi-transparent bars so both distributions are visible
bins = 30
plt.hist(sensor_a, bins=bins, alpha=0.5, label='Sensor A', color='tab:blue', edgecolor='k')
plt.hist(sensor_b, bins=bins, alpha=0.5, label='Sensor B', color='tab:orange', edgecolor='k')

# Add vertical dashed lines at each sensor's mean
mean_a = sensor_a.mean()
mean_b = sensor_b.mean()
plt.axvline(mean_a, color='tab:blue', linestyle='--', linewidth=2, label='Sensor A mean')
plt.axvline(mean_b, color='tab:orange', linestyle='--', linewidth=2, label='Sensor B mean')

plt.xlabel('Temperature (°C)')
plt.ylabel('Count')
plt.title('Histogram: Sensor Temperature Distributions')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
# Use 30 bins, alpha=0.5 for transparency so both distributions are visible.
# Add vertical dashed lines at each sensor's mean.
# Include a legend labeling each sensor.
```

```python
# Side-by-side box plot comparing Sensor A and Sensor B distributions.
# Label x-axis with sensor names, y-axis with "Temperature (deg C)".
# Add a horizontal dashed line at the overall mean of both sensors combined.

import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(6, 6))
# Prepare data
data = [sensor_a, sensor_b]

# Create boxplot with colored boxes and shown means
bp = plt.boxplot(data, labels=['Sensor A', 'Sensor B'], patch_artist=True, showmeans=True)
colors = ['tab:blue', 'tab:orange']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.6)
for median in bp['medians']:
    median.set(color='black', linewidth=2)
if 'means' in bp:
    for mean in bp['means']:
        mean.set(marker='D', markerfacecolor='white', markeredgecolor='black')

# Overall mean across both sensors
overall_mean = np.concatenate(data).mean()
plt.axhline(overall_mean, color='gray', linestyle='--', linewidth=1.5, label=f'Overall mean ({overall_mean:.2f} °C)')

plt.ylabel('Temperature (°C)')
plt.title('Box Plot: Sensor Temperature Distributions')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
